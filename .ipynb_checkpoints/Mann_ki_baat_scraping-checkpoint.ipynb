```python
def save_html(html, path):
    with open(path, 'wb') as f:
        f.write(html)
        
def open_html(path):
    with open(path, 'rb') as f:
        return f.read()
```

```python
with open('DS.html', 'r', encoding='utf-8') as f:
    page = f.read()

parser = BeautifulSoup(page, 'html.parser')

links = []
for i in parser.find_all('div', attrs={'class' : 'mkb-news-wrapper'}):
    links.append(i.a['href'])

for i in range(len(links)) :
    speech_response =  requests.get(links[i])
    speech_content = speech_response.content
    save_html(speech_response.content, 'All_speech/speech'+str(len(links)-i)+'.html')
    file.close()
    
    time.sleep(10)
    speech_parser = BeautifulSoup(speech_content, 'html.parser')

    content = speech_parser.select(".content-block")
```

```python
speech_date = []
speech = []

for i in range(len(links)) :
    html = open_html('All_speech/speech'+str(i+1)+'.html')
    parser = BeautifulSoup(html, 'html.parser')
    speech_text = parser.select('.left-container')
    for s in speech_text : 
        date = s.select(".share_date .date")[0].text
        speech_date.append(date)
        speech_ = parser.select(".news-bg")[0].text
        speech.append(speech_)
```

# LDA
---
### Topic modelling
Topic Modeling is a process to automatically identify topics present in a text object and to derive hidden patterns exhibited by a text corpus. Topic Models are very useful for multiple purposes, including:

- Document clustering
- Organizing large blocks of textual data
- Information retrieval from unstructured text
- Feature selection

Topic Modelling is different from rule-based text mining approaches that use regular expressions or dictionary based keyword searching techniques. It is an unsupervised approach used for finding and observing the bunch of words (called “topics”) in large clusters of texts.

### LDA - Latent Dirichlet Allocation
There are many approaches for obtaining topics from a text such as – Term Frequency and Inverse Document Frequency. NonNegative Matrix Factorization techniques. Latent Dirichlet Allocation is the most popular topic modeling technique.

```python
import pandas as pd
import re
import string
import spacy
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import sent_tokenize
import gensim
from gensim import corpora

all_speeches = pd.read_csv('mann_ki_baat.csv', encoding='UTF-8')
all_speeches = all_speeches.drop(['Unnamed: 0'], axis= 1)
all_speeches
```

```python
nlp = spacy.load('en_core_web_sm', disable=['parser','ner'])

def lemmatization(text, allowed_postags=['PROPN', 'NOUN', 'ADJ']):
    output = []
    for sent in text :
        doc = nlp(sent)
        output.append([token.lemma_.lower() for token in doc if token.pos_ in allowed_postags])
    return output
```

```python
all_speeches['Cleaned'] = all_speeches['Speech'].apply(lambda x: ' '.join([w for w in x.split() if len(w)> 3]))
all_speeches['Cleaned'] = all_speeches['Cleaned'].astype(str).str.replace('\d+', ' ')

all_speeches
```

```python
stop_words = set(stopwords.words('english'))
stop_words.update(['time', 'new', 'mr.', 'mrs.', 'ki', 'th', 'today', 'way', 'am', 'address', 'own', 'true', 'mann', 'baat', 'people', 'countryman',
                   'hindi', 'year', 'new', 'day', 'life', 'friend', 'friends', 'month', 'world', 'many', 'dr.', 'pm', 'modi', 'ji', 'pe', 'cook', 
                   'brinjal', 'st', 'nd', 'rd', 'sir', 'dear', 'morning', 'sunday', 'version', 'speech', 'sundays', 'authoritative', 'rendering',
                   'english', 'original', 'bag'])
all_tokens = []
def get_tokens(df, col) :
    for i in range(len(df)):
        tokenized_sent = sent_tokenize(df[col][i])
        tokenized = lemmatization(tokenized_sent)
        result = []
        for i in range(len(tokenized)) :
            tokens = tokenized[i]
            result.append([i for i in tokens if not i in stop_words])
        all_tokens.append(result)
    return all_tokens
```

```python
all_speeches['Tokens'] = get_tokens(all_speeches, 'Cleaned')
all_speeches
```

```python
LDA = gensim.models.ldamodel.LdaModel

all_topics = []
def get_topics(df, col) :
    for i in range(len(df)) :
        dictionary = corpora.Dictionary(df[col][i])
        doc_term_matrix = [dictionary.doc2bow(rev) for rev in df[col][i]]
        lda_model = LDA(corpus=doc_term_matrix, id2word=dictionary, num_topics=5, random_state=100, chunksize=1, passes=50, iterations=10)
        topic = lda_model.print_topics()
        all_topics.append(topic)
    return all_topics
```

```python
all_speeches['Topics'] = get_topics(all_speeches, 'Tokens')
```

```python
all_speeches.to_csv('Speeches_with_topics.csv')
```

```python
df = pd.read_csv('Datasets/Speeches_with_topics.csv')
from nltk.tokenize import sent_tokenize

# for i in range(len(all_speeches)) :
df['Tokenized_sent'] = df['Cleaned'].apply(sent_tokenize)
```

```python
df
```

```python
nlp = spacy.load('en_core_web_sm', disable=['parser','ner'])

import itertools

def lemmatization(text, allowed_postags=['PROPN', 'NOUN', 'ADJ']):
    output = []
    for sent in text :
        doc = nlp(sent)
        output.append([token.lemma_.lower() for token in doc if token.pos_ in allowed_postags])
    
    final = list(itertools.chain(*output))
    return final
```

```python
# for i in range(len(all_speeches)) :
stop_words = set(stopwords.words('english'))
stop_words.update(['time', 'new', 'mr.', 'mrs.', 'ki', 'th', 'today', 'way', 'am', 'address', 'own', 'true', 'mann', 'baat', 'people', 'countryman', 
                   'hindi', 'year', 'new', 'day', 'life', 'friend', 'friends', 'month', 'world', 'many', 'dr.', 'pm', 'modi', 'ji', 'pe', 'cook', 
                   'brinjal', 'st', 'nd', 'rd', 'sir', 'dear', 'morning', 'sunday', 'version', 'speech', 'sundays', 'authoritative', 'rendering', 
                   'english', 'original', 'bag', 'suggestion', 'next', 'step', 'numerous'])

final_tokenized = []
for i in range(len(df)) :
    tokens = lemmatization(df['Tokenized_sent'][i])
    final_tokenized.append([i for i in tokens if not i in stop_words])
```

```python
dictionary_all = corpora.Dictionary(final_tokenized)
doc_term_matrix_all = [dictionary_all.doc2bow(rev) for rev in final_tokenized]
```

```python
LDA = gensim.models.ldamodel.LdaModel

lda_model_all = LDA(corpus=doc_term_matrix_all, id2word=dictionary_all, num_topics=5, 
                random_state=100, chunksize=5, passes=50, iterations=100)
# lda_model_all.save('lda.model')
```

```python
lda_model_all.print_topics()
```

# Finding the optimal number of topics for LDA
---

```python
print('\nPerplexity: ', lda_model_all.log_perplexity(doc_term_matrix_all,total_docs=74))  # a measure of how good the model is. lower the better.

# Compute Coherence Score
from gensim.models.coherencemodel import CoherenceModel
coherence_model_lda = CoherenceModel(model=lda_model_all, texts=final_tokenized, dictionary=dictionary_all , coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('\nCoherence Score: ', coherence_lda)
```

```python
def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=3):
    """
    Compute c_v coherence for various number of topics

    Parameters:
    ----------
    dictionary : Gensim dictionary
    corpus : Gensim corpus
    texts : List of input texts
    limit : Max num of topics

    Returns:
    -------
    model_list : List of LDA topic models
    coherence_values : Coherence values corresponding to the LDA model with respective number of topics
    """
    coherence_values = []
    model_list = []
    for num_topics in range(start, limit, step):
        model = LDA(corpus=corpus, num_topics=num_topics, id2word=dictionary)
        model_list.append(model)
        coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())

    return model_list, coherence_values
# Can take a long time to run.
model_list, coherence_values = compute_coherence_values(dictionary=dictionary_all, corpus=doc_term_matrix_all, texts=final_tokenized, start=2, limit=16, step=2)
```

```python
limit = 16
start = 2
step = 2
x = range(start, limit, step)
plt.plot(x, coherence_values)
plt.xlabel("Num Topics")
plt.ylabel("Coherence score")
plt.legend(("coherence_values"), loc='best')
plt.show()
```
